```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([1, 2, 3, 4, 5])
y_train = np.array([300, 500, 700, 900, 1100])
print(x_train)
print(y_train)
```

```python
m = x_train.shape[0]
print(m)
```

```python
i = 0
x_i = x_train[i]
y_i = y_train[i]
print(x_i)
print(y_i)
```

```python
plt.scatter(x_train, y_train, marker='x', c='r')
plt.title("Housing Prices")
plt.ylabel('Price (in 1000s of dollars)')
plt.xlabel('Size (1000 sqft)')
plt.show()
```

```python
w = 200
b = 100
print(w)
print(b)
```

```python
def model_output(x, w, b):
  m = x.shape[0]
  f_wb = np.zeros(m)
  for i in range(m):
    f_wb[i] = w * x[i] + b
  return f_wb
```

```python
tmp_f_wb = model_output(x_train, w, b)
plt.plot(x_train, tmp_f_wb, c='b',label='Our Prediction')
plt.scatter(x_train, y_train, marker='x', c='r',label='Actual Values')
plt.title("Housing Prices")
plt.ylabel('Price (in 1000s of dollars)')
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```
