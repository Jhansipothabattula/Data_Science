# Probability Theory and Random Variables

**Basic Probability Concepts**
- Sample Space and Events
  
  - Sample Space: The Set of all possible outcomes of a Random experiment

  - Events: A subset of the sample space

- Conditional Probability

  - The Probability of an event A occuring, given that B has occured

- Independence

  - Two Events A and B are independent if
    
$${P(A \cap B)}=P(A).P(B)$$


```python
from itertools import product
# Sample space of a dice roll
sample_space = list(product(range(1, 7)))
# Probability of rolling an even number
even_numbers = [2, 4, 6]
probability_even = len(even_numbers) / len(sample_space)
print("Probability(Even) \n", probability_even)
```

**Random Variables**
- What are Random Variables?

  - Maps outcomes of a random experiment to numerical values

  - Types: Discrete | Continuous

- Probability Mass Function(PMF)

  - Probability distribution of a discrete random variable

- Probability Density Function(PDF)

  - Probability distribution of a continuous random variable

**Exception, Variance, and Standard Deviation**

- Exception(E[X])

  - Weighted average of a random variable's possible values

$E[X] = \sum_{i} x_i \, P(X = x_i)$ (Discrete)

$E[X] = \int_{-\infty}^{\infty} x \, f(x) \, dx$ (Discontinuous)

- Variance(Var[X])

  - Measures the spread of a random variable

$\mathrm{Var}(X) = E[X^2] - (E[X])^2$

- Standard Deviation($\sigma$X)

  - Square root of variance

$\sigma = \sqrt{\mathrm{Var}(X)}$




```python
import numpy as np
# Random Variable: dice roll
outcomes = np.array([1, 2, 3, 4, 5, 6])
probabilities = np.array([1/6]*6)
# Exception
exception = np.sum(outcomes * probabilities)
print("Exception: \n", exception)
# Vriance and standard Deviation
variance = np.sum((outcomes - exception) ** 2 * probabilities)
standard_deviation = np.sqrt(variance)
print("Variance: \n", variance)
print("Standard Deviation: \n", standard_deviation)
```

**Exercise 2 :- Create and analyze Random Variables**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform
# Discrete random variables: Dice roll
outcomes = [1, 2, 3, 4, 5, 6]
probabilities = [1/6] * 6
plt.bar(outcomes, probabilities, color = "pink", alpha = 0.7)
plt.title("Dice Roll Probabilities")
plt.xlabel("Outcome")
plt.ylabel("Probability")
plt.show()
# Continuous random variable: Uniform distributions
x = np.linspace(0, 1, 100)
pdf = uniform.pdf(x, loc = 0, scale = 1)
plt.plot(x, pdf, color = "red")
plt.title("PDF of Uniform(0, 1)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

> 1. Simulate flipping a coin 10,000 times and calculate probabilities of heads/tails

```python
import random
# Number of flips
N = 10000
# Counters
heads = 0
tails = 0
# Simulate coin flips
for _ in range(N):
    flip = random.choice(["H", "T"])
    if flip == "H":
        heads += 1
    else:
        tails += 1
# Probabilities
p_heads = heads / N
p_tails = tails / N
# Output
print("Total Flips: \n", N)
print("Heads Count: \n", heads)
print("Tails Count: \n", tails)
print("Probability of Heads: \n", p_heads)
print("Probability of Tails: \n", p_tails)
```

> 2. Compute the exception and variance of a weighted die(biased probabilities)

```python
import numpy as np
# Weighted die probabilities (must sum to 1)
p = np.array([0.10, 0.15, 0.20, 0.25, 0.15, 0.15])
# Die outcomes
x = np.array([1, 2, 3, 4, 5, 6])
# Expectation
E = np.sum(x * p)
# Variance
E_X2 = np.sum((x**2) * p)
Var = E_X2 - E**2
print("Probabilities: \n", p)
print("Expectation (E[X]): \n", E)
print("Variance Var(X): \n", Var)
```

> 3. Explore other distributions(ex:- Normal, Binomial) using python

```python
# Normal Distribution
import numpy as np
import matplotlib.pyplot as plt
# Normal distribution parameters
mu = 0      # mean
sigma = 1   # standard deviation
# Generate 10,000 samples
data = np.random.normal(mu, sigma, 10000)
# Plot
plt.hist(data, bins=40, density=True, color = "purple")
plt.title("Normal Distribution (=0, =1)")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()

# Binomial Distribution
import numpy as np
import matplotlib.pyplot as plt
# Binomial parameters
n = 10   # number of trials
p = 0.5  # probability of success
# Generate 10,000 samples
data = np.random.binomial(n, p, 10000)
# Plot
plt.hist(data, bins=11, density=True, color = "purple")
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.show()

# Poisson Distribution
data = np.random.poisson(lam=3, size=10000)
plt.hist(data, bins=15, density=True, color = "purple")
plt.title("Poisson Distribution (=3)")
plt.show()

# Uniform Distribution
data = np.random.uniform(low=0, high=1, size=10000)
plt.hist(data, bins=30, density=True, color= "purple")
plt.title("Uniform Distribution (0 to 1)")
plt.show()

# Exponential Distribution
data = np.random.exponential(scale=1, size=10000)
plt.hist(data, bins=40, density=True, color = "purple")
plt.title("Exponential Distribution (=1)")
plt.show()
```
