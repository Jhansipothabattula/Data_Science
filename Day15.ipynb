 # Probability Distributions in Machine Learning

**Visualizing Distributions and understanding their properties**

- Visualization helps understand skewness, kurtosis, and outliers

- Skewness

  - Measure of Symmetry: Positive skew | Negative skew

- Kurtosis

  - Measure of the "tailedness" of the distribution:
  
     High kurtosis|Low kurtosis

**Exercise 1 :- Plot and Explore Various Distributions**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm, binom, poisson, uniform

# Gaussian Distribution
x = np.linspace(-4, 4, 100)
plt.plot(x, norm.pdf(x, loc = 0, scale = 1), label = "Gaussian(u=0, s=1)")

# Binomial Distribution
n, p = 10, 0.5
x = np.arange(0, n + 1)
plt.bar(x, binom.pmf(x, n, p), alpha = 0.7, label = "Binomial(n=10, p=0.5)")
plt.title("Probability Distributions")
plt.legend()
plt.show()

# Poisson Distribution
lam = 3
x = np.arange(0, 10)
plt.bar(x, poisson.pmf(x, lam), alpha = 0.7, label = "Poisson(lam=3)")

# Uniform Distribution
x = np.random.uniform(low=0, high=10, size=1000)
sns.histplot(x, kde = True, label = "uniform", color = "red")
plt.title("Probability Distributions")
plt.legend()
plt.show()
```

> 1.Compare the effects of skewness and kurtosis on different datasets

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis, norm

# Create datasets with different skewness and kurtosis
np.random.seed(42)

# Skewness datasets
right_skew = np.random.exponential(scale=1, size=5000)
left_skew = -np.random.exponential(scale=1, size=5000)
normal_data = np.random.normal(0, 1, 5000)

# Kurtosis datasets
high_kurtosis = np.random.laplace(0, 1, 5000)       # heavy tails
low_kurtosis = np.random.uniform(-2, 2, 5000)       # flat distribution

# PLOTS
fig, axs = plt.subplots(2, 3, figsize=(16, 10))

# Skewness Plots
axs[0,0].hist(normal_data, bins=40)
axs[0,0].set_title(f"Normal (Skew={skew(normal_data):.2f})")
axs[0,1].hist(right_skew, bins=40)
axs[0,1].set_title(f"Right Skew (Skew={skew(right_skew):.2f})")
axs[0,2].hist(left_skew, bins=40)
axs[0,2].set_title(f"Left Skew (Skew={skew(left_skew):.2f})")

# Kurtosis Plots
axs[1,0].hist(normal_data, bins=40)
axs[1,0].set_title(f"Normal (Kurt={kurtosis(normal_data):.2f})")
axs[1,1].hist(high_kurtosis, bins=40)
axs[1,1].set_title(f"High Kurtosis (Kurt={kurtosis(high_kurtosis):.2f})")
axs[1,2].hist(low_kurtosis, bins=40)
axs[1,2].set_title(f"Low Kurtosis (Kurt={kurtosis(low_kurtosis):.2f})")
plt.tight_layout()
plt.show()
```
