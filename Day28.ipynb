# Introduction to Feature Engineering

**Importance of Feature Engineering in Machine Learning**

- What is Feature Engineering?

  - Process of transformimg raw data into meaningful inputs for Machine Learning models

- why is Feature Engineering important?

  - Improves Model Accuracy

  - Reduces Model Complexity

  - Enables Model Interpretability

  - Handles Data Challenges

- Key Applications

  - Finance|Healthcare|E-Commerce

**TYpes of Features: Categorical, Numerical, Ordinal**

- Categorical Features

  - Represent discrete categories or labels

  - Encoding Techniques

    - One-Hot Encoding

    - Label encoding

- Numerical Features

  - Represent Continuous or discrete numbers

  - Preprocessing techniques

    - Scaling

- Ordinal Features

  - Represent categorical data with a meaningful order

**Overview of Feature Engineering Techniques**

- Scaling

  - Ensures all Features contribute equally to the model

  - Techniques: Min-Max Scaling|Standardization

- Encoding

  - Converts categorical data into Numerical Format

  Techniques: One-Hot Encoding|Label Encoding

- Transformation

  - Applies mathematical functions to modify features

  - Examples: Log Transformation|Polynomial Features

- Feature Selection

  - Reduces the number of input features to improve model perfomance

  - Techniques: Statistical Methods|Recurisve Feature elimination(RFE)

```python
import pandas as pd

# Load Titanic Dataset
from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv("titanic.csv")
print(df.info())

# Preview the first few rows
print("Dataset preview: \n")
print(df.head())

# Seperate Features
categorical_features = df.select_dtypes(include=["object"]).columns
numerical_features = df.select_dtypes(include=["int64"]).columns

print("Categorical Features: ", categorical_features.tolist())
print("Numerical Features: ", numerical_features.tolist())

# Display Summary of Categorical Features
print("Summary of Categorical Features: \n")
for col in categorical_features:
  print(f"{col}:\n", df[col].value_counts(), "\n")

# Display Summary of Numerical Features
print("Numerical Features:")
print(df[numerical_features].describe())
```
